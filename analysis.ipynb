```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
with open("gossip_wide_10_non_iid2.json") as f:
    gossip_data = json.load(f)

with open("mutual_wide_10_before.json") as f:
    mutual_data = json.load(f)

with open("mutual_wide_10_non_iid2.json") as f:
    mutualafter_data = json.load(f)

with open("collaborative_wide_10_non_iid2.json") as f:
    collaborative_data = json.load(f)
```

```python
gossip_key = []
gossip_value = []
mutual_key =[]
mutual_value =[]
mutualafter_key =[]
mutualafter_value =[]
collaborative_key =[]
collaborative_value =[]

for i in gossip_data:
    keys = list(i.keys())
    values = list(i.values())
    gossip_key.append(keys[0])
    gossip_value.append(values[0])

for i in mutual_data:
    keys = list(i.keys())
    values = list(i.values())
    mutual_key.append(keys[0])
    mutual_value.append(values[0])

for i in mutualafter_data:
    keys = list(i.keys())
    values = list(i.values())
    mutualafter_key.append(keys[0])
    mutualafter_value.append(values[0])

for i in collaborative_data:
    keys = list(i.keys())
    values = list(i.values())
    collaborative_key.append(keys[0])
    collaborative_value.append(values[0])
```

```python
plt.plot(gossip_key,gossip_value,label="gossip")
# plt.plot(mutual_key,mutual_value,label="Def-KT")
plt.plot(mutualafter_key,mutualafter_value,label="Def-KT")
plt.plot(collaborative_key,collaborative_value,label="Collaborative")
plt.legend()
```
